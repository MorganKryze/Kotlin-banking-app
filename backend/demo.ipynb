```python
import cv2
import imutils
import matplotlib.pyplot as plt

# Load the original image
img_orig = cv2.imread("../../test-image.jpg")
image = img_orig.copy()
image = imutils.resize(image, width=500)
ratio = img_orig.shape[1] / float(image.shape[1])

# Convert the image to grayscale, blur it slightly, and then apply edge detection
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
edged = cv2.Canny(blurred, 75, 200)

# Find contours in the edge map and sort them by size in descending order
cnts = cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cnts = imutils.grab_contours(cnts)
cnts = sorted(cnts, key=cv2.contourArea, reverse=True)

# Initialize a contour that corresponds to the receipt outline
receiptCnt = None

# Loop over the contours to find the receipt outline
for c in cnts:
    # Approximate the contour
    peri = cv2.arcLength(c, True)
    approx = cv2.approxPolyDP(c, 0.02 * peri, True)

    # If the approximated contour has four points, we have found the receipt outline
    if len(approx) == 4:
        receiptCnt = approx
        break

# Check if we found the receipt contour
if receiptCnt is None:
    raise Exception(

            "Could not find receipt outline. "
            "Try debugging your edge detection and contour steps.",

    )

# Apply a four-point perspective transform to obtain a top-down view of the receipt
def four_point_transform(image, pts):
    rect = np.zeros((4, 2), dtype="float32")

    # top-left point
    rect[0] = pts[0]
    # top-right point
    rect[1] = pts[1]
    # bottom-right point
    rect[2] = pts[2]
    # bottom-left point
    rect[3] = pts[3]

    # order the points in the rectangle
    (tl, tr, br, bl) = rect

    # compute the width of the new image
    widthA = np.linalg.norm(br - bl)
    widthB = np.linalg.norm(tr - tl)
    maxWidth = max(int(widthA), int(widthB))

    # compute the height of the new image
    heightA = np.linalg.norm(tr - br)
    heightB = np.linalg.norm(tl - bl)
    maxHeight = max(int(heightA), int(heightB))

    # destination points for the perspective transform
    dst = np.array([
        [0, 0],
        [maxWidth - 1, 0],
        [maxWidth - 1, maxHeight - 1],
        [0, maxHeight - 1],
    ], dtype="float32")

    # perform the perspective transform and return the result
    M = cv2.getPerspectiveTransform(rect, dst)
    warped = cv2.warpPerspective(image, M, (maxWidth, maxHeight))

    return warped

# Apply the transformation to get the receipt
receipt = four_point_transform(img_orig, receiptCnt.reshape(4, 2) * ratio)

# Display the results using matplotlib
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Detected Receipt Outline")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(receipt, cv2.COLOR_BGR2RGB))
plt.title("Top-Down View of Receipt")
plt.axis("off")

plt.show()
```

```python
import os

import cv2
import imutils
import matplotlib.pyplot as plt


class Receipt:

    def __init__(self, image_name, output_dir="../../data/img/capture"):
        # Define the base directory
        self.base_dir = "data/img/capture"

        # Construct the full image path
        image_path = os.path.join(self.base_dir, image_name)

        self.img_orig = cv2.imread(image_path)

        # Check if the image was loaded successfully
        if self.img_orig is None:
            raise ValueError(f"Could not load image at path: {image_path}. Please check the file path.")

        self.image = imutils.resize(self.img_orig, width=500)
        self.ratio = self.img_orig.shape[1] / float(self.image.shape[1])
        self.receiptCnt = None
        self.receipt = None
        self.output_dir = output_dir # Set the output directory

    def edge_detection(self):
        gray = cv2.cvtColor(self.image, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        return cv2.Canny(blurred, 75, 200)

    def color_thresholding(self):
        hsv = cv2.cvtColor(self.image, cv2.COLOR_BGR2HSV)
        lower_bound = np.array([0, 0, 200])
        upper_bound = np.array([255, 255, 255])
        mask = cv2.inRange(hsv, lower_bound, upper_bound)

        kernel = np.ones((5, 5), np.uint8)
        return cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

    def find_receipt_contour(self):
        edged = self.edge_detection()
        mask = self.color_thresholding()

        cnts = imutils.grab_contours(
            cv2.findContours(edged.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE),
        ) + imutils.grab_contours(
            cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE),
        )

        max_area = 0
        for c in cnts:
            peri = cv2.arcLength(c, True)
            approx = cv2.approxPolyDP(c, 0.02 * peri, True)

            if len(approx) == 4:
                area = cv2.contourArea(c)
                if area > max_area:
                    max_area = area
                    self.receiptCnt = approx

        if self.receiptCnt is None:
            raise Exception("Could not find receipt outline.")

    def order_points(self, pts):
        rect = np.zeros((4, 2), dtype="float32")
        s = pts.sum(axis=1)
        rect[0] = pts[np.argmin(s)]  # Top-left
        rect[2] = pts[np.argmax(s)]  # Bottom-right

        diff = np.diff(pts, axis=1)
        rect[1] = pts[np.argmin(diff)]  # Top-right
        rect[3] = pts[np.argmax(diff)]  # Bottom-left

        return rect

    def four_point_transform(self, pts):
        rect = self.order_points(pts)
        (tl, tr, br, bl) = rect

        widthA = np.linalg.norm(br - bl)
        widthB = np.linalg.norm(tr - tl)
        maxWidth = max(int(widthA), int(widthB))

        heightA = np.linalg.norm(tr - br)
        heightB = np.linalg.norm(tl - bl)
        maxHeight = max(int(heightA), int(heightB))

        dst = np.array([
            [0, 0],
            [maxWidth - 1, 0],
            [maxWidth - 1, maxHeight - 1],
            [0, maxHeight - 1],
        ], dtype="float32")

        M = cv2.getPerspectiveTransform(rect, dst)
        return cv2.warpPerspective(self.img_orig, M, (maxWidth, maxHeight))

    def process_receipt(self):
        self.find_receipt_contour()
        self.receipt = self.four_point_transform(self.receiptCnt.reshape(4, 2) * self.ratio)

    def display_results(self):
        plt.figure(figsize=(10, 10))
        plt.subplot(1, 2, 1)
        plt.imshow(cv2.cvtColor(self.image, cv2.COLOR_BGR2RGB))
        plt.title("Detected Receipt Outline")
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(cv2.cvtColor(self.receipt, cv2.COLOR_BGR2RGB))
        plt.title("Top-Down View of Receipt")
        plt.axis("off")

        plt.show()

    def save_receipt_image(self):
        if self.receipt is not None:
            # Create the output directory if it doesn't exist
            os.makedirs(self.output_dir, exist_ok=True)
            output_path = os.path.join(self.output_dir, "processed_receipt.jpg")
            cv2.imwrite(output_path, self.receipt)
            print(f"Receipt image saved to {output_path}")
        else:
            print("No receipt image to save. Please process the receipt first.")
```

```python
# Example usage
try:
    receipt = Receipt("../../data/img/capture/test-image.jpg")  # Use the correct image file path
    receipt.process_receipt()
    receipt.display_results()
    receipt.save_receipt_image()  # Saves to the default directory
except ValueError as e:
    print(e)
except Exception as e:
    print(e)  # Catch any other exceptions that might occur
```

```python
import re

import pytesseract
from PIL import Image

# Load the image from file
image_path = "../../test-image-cleaned.jpg"  # Path to your image
image = Image.open(image_path)

# Perform OCR
extracted_text = pytesseract.image_to_string(image)

# Clean up the text
lines = extracted_text.splitlines()
useful_lines = [line for line in lines if line.strip()]

# Print the list of lines found by Tesseract
print("Lines found by Tesseract:")
for line in useful_lines:
    print(line)

# Find amount after "MONTANT"
match = re.search(r"MONTANT\s*:\s*([\d,.]+)\s*EUR", "\n".join(useful_lines))

if match:
    amount_str = match.group(1)  # Keep the original format (with comma)
    print(f'Amount after "MONTANT": {amount_str} EUR')
else:
    print('No amount found after "MONTANT".')
```
